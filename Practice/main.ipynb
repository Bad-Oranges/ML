```python
train_data = data[data['中标时间'].dt.year >= 2020].copy()
train_data['year'] = train_data['中标时间'].dt.year
train_data['month'] = train_data['中标时间'].dt.month
result = train_data.groupby(['year', 'month']).agg(count=('标段名称','size'), value=('中标价','sum'))
result
```

## 定义/训练模型

```python
import torch
import torch.nn as nn
import numpy as np

# 数据归一化
normalized_data = (dataset - dataset.min(axis=0)) / (dataset.max(axis=0) - dataset.min(axis=0))

# 定义超参数
input_size = 2  # 输入特征的大小，这里是 count 和 value，所以为 2
hidden_size = 64  # 隐层的大小
output_size = 2  # 输出的大小，这里是预测 count 和 value，所以为 2
num_epochs = 5000  # 迭代次数
learning_rate = 0.0001  # 学习率

# RNN和数据集参数
time_steps = 6
train_seq_length = 24

# 生成训练数据和目标数据
train_data = normalized_data[:train_seq_length]
test_data = normalized_data[train_seq_length:]

# print(train_data.shape, test_data.shape)

def generate_sequences(_data, time_steps):
    sequences = []
    targets = []
    for i in range(len(_data) - time_steps):
        sequences.append(_data[i:i+time_steps])
        targets.append(_data[i+time_steps])
    return torch.tensor(sequences).float(), torch.tensor(targets).float()


train_sequences, train_targets = generate_sequences(train_data, time_steps)
test_sequences, test_targets = generate_sequences(test_data, time_steps)


# 定义RNN模型
class RNNModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(RNNModel, self).__init__()
        self.hidden_size = hidden_size
        self.rnn = nn.RNN(input_size, hidden_size, batch_first=True)
        self.out = nn.Sequential(
            nn.BatchNorm1d(hidden_size),
            nn.LeakyReLU(True),
            nn.Linear(hidden_size,hidden_size),
            nn.Linear(hidden_size, output_size),
        )

    def forward(self, x):
        _, h = self.rnn(x)
        out = self.out(h.squeeze(0))
        return out


# 创建模型实例
model = RNNModel(input_size, hidden_size, output_size)

# 定义损失函数和优化器
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

# 训练模型
for epoch in range(num_epochs):
    outputs = model(train_sequences)
    loss = criterion(outputs, train_targets)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch+1) % 500 == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item()}")


# 测试模型
model.eval()
with torch.no_grad():
    test_outputs = model(test_sequences)
    test_loss = criterion(test_outputs, test_targets)
    print(f"Test Loss: {test_loss.item()}")
```

## 机器学习算法说明

机器学习算法是一类用于从数据中自动学习模式和规律的算法。这些算法使得计算机能够通过对大量数据的分析和学习来做出预测、分类、聚类等任务，而无需明确的程序指令。

在这个作业中使用了监督学习（Supervised Learning）的技术。在监督学习中，算法通过已知输入和对应的输出数据（标签）进行训练，从而学习输入与输出之间的映射关系。常见的监督学习算法包括线性回归、逻辑回归、决策树、支持向量机（SVM）和我们这里用到的神经网络等。

神经网络是一种计算模型，受到人类神经系统的启发而设计。它由大量的人工神经元（或称为节点）组成，这些神经元通过连接权重相互连接。神经网络能够从输入数据中学习并进行模式识别、分类和预测等任务。

神经网络的基本包括以下几个方面：

神经元（节点）：神经网络中的基本单元，接收输入信号，并通过激活函数处理这些信号，产生输出结果。
权重：连接神经元之间的权重表示它们之间的连接强度，权重值越大，对应连接的重要性越高。
偏置：每个神经元都有一个偏置，用于调整激活函数的输入。
激活函数：激活函数将神经元的加权输入转换为输出结果。常见的激活函数包括sigmoid、ReLU（Rectified Linear Unit）等。
前向传播：从输入层开始，通过对每个神经元进行加权求和和激活函数操作，将信号传递到下一层。
反向传播：通过计算输出结果与预期结果之间的误差，并将误差从输出层向前传播，调整连接权重和偏置，以优化神经网络的性能。


循环神经网络（Recurrent Neural Network，RNN）是一种具有内部循环结构的神经网络。RNN的主要特点是能够处理序列数据，如语言、时间序列等。相比于传统神经网络，RNN能够利用之前的信息来影响当前的输出，具有记忆能力。

RNN的内部结构包括以下几个关键部分：

输入层：接收序列数据作为输入。

隐藏层：包含了循环连接，使得网络能够在处理当前输入的同时，保留之前的状态信息。

输出层：生成RNN的输出结果。

循环连接：将隐藏层的输出反馈到自身或下一个时间步，使得网络能够保持对序列数据的记忆。

时间步：RNN中的每一次计算被认为是一个时间步，每个时间步处理输入序列的一个元素。

RNN通过时间的展开来处理序列数据，展开后的网络结构可以看作是多个相同网络的堆叠，每个时间步共享相同的权重参数。这样，RNN能够在处理序列时考虑到前面时间步的信息，从而更好地捕捉序列中的模式和依赖关系。

在实际应用中，RNN还有一些变种，如长短期记忆网络（Long Short-Term Memory，LSTM）和门控循环单元（Gated Recurrent Unit，GRU），它们通过引入门控机制来解决传统RNN存在的梯度消失和梯度爆炸等问题，提高了网络的学习能力和效果。
